```python
import matplotlib.pyplot as plt

# Convert table data into a dictionary
data = {
    "DBSCAN": {"Precision": 85.09, "Recall": 51.53, "F Score": 55.735, "Conditional Entropy": 1.619},
    "Spectral Clustering": {"Precision": 64.5, "Recall": 69.4, "F Score": 63.5, "Conditional Entropy": 1.221},
    "K-means": {"Precision": 47.31, "Recall": 84.24, "F Score": 56.68, "Conditional Entropy": 1.65}
}

# Create a separate plot for each metric
for metric in ["Precision", "Recall", "F Score", "Conditional Entropy"]:
    plt.figure(figsize=(10, 5))
    values = [values[metric] for values in data.values()]
    bars = plt.bar(data.keys(), values)
    
    # Annotate the metric value in the center of each bar
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval/2, round(yval, 2), ha='center', va='center', color='white', fontweight='bold', fontsize=14)
    
    plt.title(metric)
    plt.show()
```
